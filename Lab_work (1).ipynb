````python
sns.histplot(df_all['G3'],kde=True)
plt.title("Distribution of Final Grade")
plt.show()
````

CHECKING THE DISTRBUTION OF 2 OR MORE VARIABLES

1.KINETIC DENSITY ESTIMATION(KDE PLOT)

````python
#probability distribution of numerical variable
sns.displot(df_all['G3'], kde=True)
plt.title("Distribution of Final Grade (using displot)")
plt.show()
````

````python
#box plot sex vs studytime
#they show /display the distibution of numerical variable accross different categories of one or more categorical variables.
sns.boxplot(x="sex", y="studytime",data= df_all)
plt.title("Distribution of Final Grade (using boxplot)")
plt.show()
````

````python
plt.hist(df_all['G3'], bins=20) # You can adjust the number of bins
plt.title("Distribution of Final Grade (using matplotlib hist)")
plt.xlabel("G3")
plt.ylabel("Frequency")
plt.show()
````

BOXPLOT

````python
sns.boxplot(x= "studytime",y="G3",data= df_all)
plt.title("How studytime is affecting the final grade (using boxplot)")
plt.show()
````

VIOLINPLOT

````python
sns.violinplot(x="studytime", y= "G3", data= df_all)
plt.title("How studytime is affecting final grade" (using violinplot)")
plt.show()
````

````python
sns.violinplot(x="sex", y= "G3", data= df_all)
plt.title("How sex is affecting final grade (using violinplot)")
plt.show()
````

COUNTPLOT

````python
sns.countplot(x="sex", data= df_all , hue = 'sex')
plt.title("How sex is affecting final grade (using countplot)")
plt.show()
````

````python
sns.countplot(x="studytime", data= df_all ,hue = "studytime")
plt.title("How studytime is affecting final grade (using countplot)")
plt.show()
````

````python
sns.countplot(x="address", data= df_all, hue = 'address')
plt.title("How address is affecting final grade (using countplot)")
plt.show()
````

````python


# Assuming df_all is already loaded and processed as in the previous cells

# Count the occurrences of each category in the 'sex' column
sex_counts = df_all['sex'].value_counts()

# Create the pie chart
plt.figure(figsize=(6, 6)) # Set the figure size
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
# sex_counts: The values for each wedge of the pie
# labels: The labels for each wedge (the categories from sex_counts.index)
# autopct: Formats the percentage displayed on each wedge
# startangle: Starts the first wedge at the top (90 degrees)
# colors: Optional list of colors for the wedges

plt.title("Distribution of Sex") # Set the title of the pie chart
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````
