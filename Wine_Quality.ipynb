DATA VISUALIZATION

HISTOGRAM

```python
sns.histplot(df['quality'],kde=True)
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.show()
```

BOX PLOT

```python
sns.boxplot(x=df['quality'])
plt.title('Box Plot of Wine Quality')
plt.xlabel('quality')
plt.ylabel('density')
plt.show()
```

```python
sns.boxplot(x=df['quality'])
plt.title('How ph is affecting quality')
plt.xlabel('quality')
plt.ylabel('ph')
plt.show()
```

```python
sns.boxplot(x=df['quality'])
plt.title('How alcohol is affecting quality')
plt.xlabel('quality')
plt.ylabel('alcohol')
plt.show()
```

VIOLIN PLOT

```python
sns.violinplot(x=df['quality'])
plt.title('Violin Plot of Wine Quality')
plt.xlabel('quality')
plt.ylabel('density')
plt.show()
```

```python
sns.violinplot(x=df['quality'])
plt.title('How ph is affecting quality')
plt.xlabel('quality')
plt.ylabel('ph')
plt.show()
```

```python
sns.violinplot(x=df['quality'])
plt.title('How alcohol is affecting quality')
plt.xlabel('quality')
plt.ylabel('alcohol')
plt.show()
```

COUNT PLOT

```python
sns.countplot(x=df['quality'])
plt.title('Count Plot of Wine Quality')
plt.xlabel('quality')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x=df['quality'])
plt.title('How ph is affecting quality')
plt.xlabel('quality')
plt.ylabel('ph')
plt.show()
```

CORRELATION MATRIX-heatmap

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.pie(df['quality'].value_counts(), labels=df['quality'].unique(), autopct='%1.1f%%')
plt.title('Distribution of Wine Quality')
plt.show()
```
